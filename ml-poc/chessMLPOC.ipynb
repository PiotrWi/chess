```python
data, evaluation = DataAccess.get_all_evaluated_records()
```

```python
print(data)
print(data.shape)
```

```python
data = [DataAccess.fen_positon_to_one_hot_vector(fen) for fen in data]
```

```python
data = np.array(data)
print(data.shape)
```

```python
import tensorflow as tf
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1000, activation="relu"),
    tf.keras.layers.Dense(300, activation="relu"),
    tf.keras.layers.Dense(1, activation="relu")])
```

```python
model.compile(loss="mean_absolute_error", optimizer="sgd", metrics=["mean_absolute_error"])
```

```python
history = model.fit(data, evaluation, epochs=1000)
```
