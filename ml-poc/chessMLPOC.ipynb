```python
import os
import numpy as np
import pandas as pd
import sklearn.model_selection
import tensorflow as tf

import EvaluationDataAccess as DataAccess
```

```python
print(os.getcwd())
```

```python
X, Y = DataAccess.get_all_evaluated_records()
X = np.array([DataAccess.fen_positon_to_one_hot_vector(fen) for fen in X])

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, Y, test_size=0.33)
```

```python
print(X_train)
```

```python
print(X.shape)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(500, activation="relu"),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(1, activation="relu")])

optimizer = tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9, nesterov=True)
model.compile(loss="mean_squared_logarithmic_error", optimizer=optimizer, metrics=["mean_squared_logarithmic_error"])

history = model.fit(X_train, y_train, epochs=150, validation_data=(X_test, y_test))
```

```python
model.save("chess500r100r1r.tf", save_format="tf")
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(500, activation="relu"),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1, activation="relu")])

optimizer = tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9, nesterov=True)
model.compile(loss="mean_squared_logarithmic_error", optimizer=optimizer, metrics=["mean_squared_logarithmic_error"])

history = model.fit(X_train, y_train, epochs=150, validation_data=(X_test, y_test))
```

```python
model.save("chess500r100r1rDO.tf", save_format="tf")
```
